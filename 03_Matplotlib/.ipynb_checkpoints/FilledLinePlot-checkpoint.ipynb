```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
ages_x = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43,
          45, 47, 49, 51, 53, 55, 57, 59, 61, 63]
```

```python
dev_y = [
    43000, 46000, 49000, 52000, 55000,
    58000, 61000, 64000, 67000, 70000,
    73000, 76000, 79000, 82000, 85000,
    88000, 91000, 94000, 97000, 100000
]
py_dev_y = [45000, 47600, 54300, 56000, 60000,
            64000, 68000, 53000, 49600, 51200,
            83000, 86000, 89000, 97000, 95000,
            98000, 101000, 104000, 107000, 110000]
js_dev_y = [32600, 39300, 48000, 51000, 54000,
            57000, 60000, 61200, 66000, 69000,
            72000, 75000, 78000, 79200, 84000,
            87000, 90000, 93000, 96000, 99000]

ages_x = np.array(ages_x)
py_dev_y = np.array(py_dev_y)
js_dev_y = np.array(js_dev_y)
```

```python
plt.plot(ages_x,py_dev_y,label = 'Python Dev')
plt.plot(ages_x,js_dev_y,color = "black", linestyle = "--", label = 'Js Dev')

plt.fill_between(ages_x,py_dev_y,alpha = 0.25)

plt.legend()
plt.xlabel("Ages")
plt.ylabel("Median Salary(USD)")
plt.title("Median Salary(USD) by Age")
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages_x,py_dev_y,label = 'Python Dev')
plt.plot(ages_x,js_dev_y,color = "black", linestyle = "--", label = 'Js Dev')

plt.fill_between(ages_x,py_dev_y,js_dev_y
                 ,where = (py_dev_y > js_dev_y)
                 ,interpolate = True,alpha = 0.25,label = "Above Avg")

plt.fill_between(ages_x,py_dev_y,js_dev_y,
                 where = (py_dev_y <= js_dev_y)
                 ,interpolate = True
                 ,color = "red", alpha = 0.25,label = "Beloe Avg")

plt.legend()
plt.tight_layout()
plt.xlabel("Ages")
plt.ylabel("Median Salary(USD)")
plt.title("Median Salary(USD) by Age")
plt.savefig('FilledLinePlot.png')
plt.show()
```
