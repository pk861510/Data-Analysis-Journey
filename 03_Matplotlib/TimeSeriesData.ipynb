```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
df = pd.read_csv("D:\\Data Analytics\\Data\\swiggy.csv")
```

```python
df.head(2)
```

```python
df["DateTime"] = pd.to_datetime(df["DateTime"])
df.sort_values("DateTime",inplace = True)
```

```python
date = df["DateTime"]
price = df["SWIGGYEQN"]
```

```python
plt.figure(figsize =(12,6))
plt.plot_date(date,price,linestyle = "solid")
plt.gcf().autofmt_xdate()

# Formate Date using matplotlibe dates
date_format = mpl_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)

plt.xlabel("Date")
plt.ylabel("Swiggy Price")
plt.title("Monthly Swiggy Closing Price using Matplotlib")
plt.tight_layout()
plt.savefig("SwiggyMonthly.png")
plt.show()
```
