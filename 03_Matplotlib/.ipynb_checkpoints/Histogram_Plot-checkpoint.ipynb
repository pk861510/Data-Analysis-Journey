```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
age_grp = [10,15,45,65,63,23,24,49,34,39,29,45,69,55,34,24]
ages = np.array(age_grp)
median_age = np.median(ages)
# print(median_age)
```

```python
bins = [10,20,30,40,50,60,70]
```

```python
plt.hist(ages,bins = bins,edgecolor= 'black')

plt.axvline(median_age,color="red",label = "Median Age")
plt.legend()
plt.xlabel("Ages")
plt.ylabel("Count of Ages")
plt.savefig('histogran_ages.png')
plt.show()
```
