```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
# generate the random 20 No for scatter plot using numpy rnadom no generator
rng = np.random.default_rng(seed = 1)
x = rng.integers(40,160,20)
y = rng.integers(60,180,20)
# color could be anything rating of product the satisfied
colors = rng.integers(1,10,20)
# used this size to the actual size of our data point
sizes = rng.integers(285,500,20)
```

```python
# Set the style for our scatter plot plot
# plt.style.available
plt.style.use('seaborn-v0_8-dark-palette')
```

```python
plt.scatter(x,y,s = sizes,c=colors,cmap = 'Blues',alpha = 0.75,edgecolor = 'black',linewidth = 1)

cbar = plt.colorbar()
cbar.set_label("Satisfaction")
plt.title("Scatter Plot Using Matplotlib")
plt.tight_layout()
plt.show()
```
