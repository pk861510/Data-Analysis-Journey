```python
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter
```

```python
pd.set_option('display.max_columns',139)
```

```python
df = pd.read_csv("D:\\Data Analytics\\Data\\survey_results_public.csv")
```

```python
lang = df['LanguageHaveWorkedWith'].dropna()
# lang
```

```python
# Count the langeage from the Survey
counter = Counter()
for item in lang:
    counter.update(item.split(';'))
```

```python
counter.most_common(5)
```

```python
# Extract the language and their count in seperate list
language = []
count = []
for item in counter.most_common(5):
    language.append(item[0])
    count.append(item[1])
explode = [0,0,0,0.1,0]
```

```python
# Plot the pie chart based on the survey data
plt.pie( count,labels= language, autopct = "%1.1f%%", shadow = True
        ,startangle = 90, wedgeprops = {'edgecolor': 'black'} 
       , explode = explode)

plt.tight_layout()
plt.title("Most Used Language")
plt.savefig('Pie_Chart1.png')
plt.show()
```
