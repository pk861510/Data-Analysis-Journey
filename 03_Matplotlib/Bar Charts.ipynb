```python
data = pd.read_csv("data.csv")
```

```python
data.head(2)
```

```python
id = data["Responder_id"]
languages = data["LanguagesWorkedWith"]
```

```python
language_counter = Counter()
language_count = []
language = []
```

```python
for item in languages:
    language_counter.update(item.split(";"))
```

```python
# language_counter.most_common(13)
```

```python
for item in language_counter.most_common(15):
    language.append(item[0])
    language_count.append(item[1])
```

```python

plt.barh(language,language_count)
plt.ylabel("Programming Languages")
plt.title("Most Popular Languages")
plt.savefig("bar_plot_data.png")
plt.show()
```

```python
language_count.reverse()
language.reverse()
```

```python
plt.figure(figsize=(12,6)) 
plt.barh(language,language_count)
plt.ylabel("Programming Languages")
plt.title("Most Popular Languages")
plt.savefig("bar_plot_data.png")
plt.show()
```
