```python
import requests
import json
import pandas as pd
import time
```

```python
ls_no = []
session_no = []
api_data = []
url = []
```

```python
# https://sansad.in/api_ls/member/getMemberAttendanceMonthWise?loksabha=18&session=1&locale=en
BASE_URL = "https://sansad.in/api_ls/member"
LOCK_SABHA_NO = 15
while LOCK_SABHA_NO <= 18:
    SESSION_NO = 1
    # print(LOCK_SABHA_NO)
    while SESSION_NO <=20:
        session_url = F"{BASE_URL}/getMemberAttendanceMonthWise?loksabha={LOCK_SABHA_NO}&session={SESSION_NO}&locale=en"
        # print(session_url)
        try:
            session_response = requests.get(session_url)
            session_response.raise_for_status()
            sessions = session_response.json()
            if len(sessions) == 0:
                break
            # print(sessions)
            ls_no.append(LOCK_SABHA_NO)
            session_no.append(SESSION_NO)
            api_data.append(sessions)
            url.append(session_url)
            SESSION_NO += 1
        except request.exception.RequestException as e:
            print({e})
    LOCK_SABHA_NO += 1
    time.sleep(1)
```

```python
ls_data = pd.DataFrame(
    {"lok_sabha":ls_no,
     "session":session_no,
     "api_data": api_data,
     "url": url}
)
     
```

```python
ls_data["api_data"].head(5)
```

```python
ls_data["api_data"][1]
# explode data for one data in one multiple rows
```

```python
exploded_df = ls_data.explode('api_data').reset_index().drop(columns=["index"])
# Extract Date of attendance and convert it to Date time
exploded_df["dateOfAttendance"] = exploded_df["api_data"].apply(lambda x: x["dateOfAttendance"])
exploded_df["dateOfAttendance"] = pd.to_datetime(exploded_df["dateOfAttendance"])
# formate 
exploded_df["dateOfAttendance"] = exploded_df["dateOfAttendance"].dt.strftime("%Y/%m/%d")
```

```python
member_df = exploded_df.copy()
```

```python
member_df.head(10)
```

```python
attendance_rows= []

for index,row in member_df.iterrows():
    lok_sabha_num = row["lok_sabha"]
    session_no = row["session"]
    attendance_date = row["dateOfAttendance"]
    print(f"Lok_sabha:{lok_sabha_num},Sessions:{session_no},Date:{attendance_date}")
    session_url = f"{BASE_URL}/getMemberAttendanceDateWise?loksabha={lok_sabha_num}&session={session_no}&dateOfAttendance={attendance_date}&locale=en"
    session_response = requests.get(session_url)
    data = session_response.json()
    attendance_rows.append({
        "lok_sabha": lok_sabha_num,
        "session": session_no,
        "dateOfAttendance": attendance_date,
        "member_level_attendance": data   # entire list stored in one column
    })
    time.sleep(1)
```

```python
df = pd.DataFrame(attendance_rows)
```

```python
df.to_csv("attendance.csv")
```
